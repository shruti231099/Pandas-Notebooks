```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

match=pd.read_csv('matches - matches.csv')
delivery=pd.read_csv('deliveries.csv')

ipl=delivery.merge(match,left_on='match_id',right_on='id')

def season_performance(batsman_name):
    
    single=ipl[ipl['batsman']==batsman_name]
    performance=single.groupby('season')['batsman_runs'].sum().reset_index()
   
    #converting dataframe column to a list
    season=performance['season'].tolist()
    batsman_run=performance['batsman_runs'].tolist()
   
    #returning graph
    plt.ylabel("Batsman_Runs")
    plt.xlabel("Year")
    plt.title("Bataman Season by Season Performance")
    plt.plot(season,batsman_run, color='green')
    return plt.show()
```

```python
season_performance('V Kohli')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

match=pd.read_csv('matches - matches.csv')

delivery=pd.read_csv('deliveries.csv')

#merging two datasets
ipl=delivery.merge(match,left_on='match_id',right_on='id')



def runs_scored(batsman_name):
    
    single=ipl[ipl['batsman']==batsman_name]
    data=single.groupby('batsman_runs')['ball'].count().reset_index()
  
    #converting dataframe column to a list
    ball=data['ball'].tolist() 
    batsman_runs=data['batsman_runs'].tolist()
    
    #ploting pie chart
    plt.pie(ball,labels=batsman_runs,autopct='%1.1f%%',radius=2,explode=(0.1,0.1,0.1,0.3,0.1,0.1))
    return plt.show()
```

```python
runs_scored('CH Gayle')
```
